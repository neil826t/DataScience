```python
elections = elections.dropna()
```

```python
elections['FIPS'] = elections.county_fips.astype(int)
```

```python
elections = elections.drop('office', axis=1)
```

```python
elections = elections[elections['mode']=='TOTAL'].drop('mode',axis=1)
```

```python
elections = elections[elections.party.isin(['DEMOCRAT','REPUBLICAN'])]
```

```python
el2012 = elections[(elections.year == 2012) & (elections['mode'] == 'TOTAL')]
```

```python
el2012win = el2012[el2012.groupby(['county_fips'])['candidatevotes'].transform(max) == el2012['candidatevotes']]
```

```python
el2012win = el2012win.drop_duplicates('county_fips')
```

```python
el2012win.to_parquet('election_winners_2012_county.parquet')
```

```python
el2012win['fips'] = el2012win.county_fips.astype(int)
```

```python
el2012win_basic = el2012win[['fips','party']]
```

```python
counties_with_2012_party = pd.merge(counties, el2012win_basic, left_on='fips', right_on='fips').rename(columns={'party':'winning_party_2012'})
```

```python
counties_with_2012_party.to_parquet('counties_with_2012_winning_party.parquet')
```
